```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cities_r2.csv")
```

```python
df.head()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
X = df[["effective_literacy_rate_total"]]
```

```python
Z = linkage(X, method="ward")
```

```python
plt.figure(figsize=(10, 5))
plt.title("Dendrogram (Hierarchical Clustering)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
dendrogram(Z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters = 3)
```

```python
hc.fit(X)
```

```python
df["Clusters"] = hc.labels_
```

```python
plt.scatter(df['effective_literacy_rate_total'], [0]*len(df), c=df['Clusters'], cmap='rainbow')
plt.title("Agglomerative Clustering Result")
plt.xlabel("Effective Literacy Rate (Total)")
plt.show()
```
