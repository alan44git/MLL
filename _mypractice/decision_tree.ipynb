```python
import pandas as pd
```

```python
df = pd.read_csv("./datasets/madfhantr.csv")
df.head()
```

```python
df = df.drop('Loan_ID',axis=1)
```

```python
df["Dependents"] = df["Dependents"].replace('3+',3).astype(float)L
```

```python
numeric_cols = df.select_dtypes(include = ['int64','float64'])
```

```python
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df = df.fillna(df.mode().iloc[0])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_cols = df.select_dtypes(include=['object'])
encodings = {}
for cols in categorical_cols:
    le = LabelEncoder()
    df[cols] = le.fit_transform(df[cols])
    encodings[cols]=le
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=62)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Classification report:\n", classification_report(y_test,y_pred))
```

```python
encodings.pop('Loan_Status')
```

```python
new_data = pd.DataFrame({
    'Gender': ['Male'],
    'Married': ['Yes'],
    'Dependents': [2.0],
    'Education': ['Graduate'],
    'Self_Employed': ['No'],
    'ApplicantIncome': [100000],
    'CoapplicantIncome': [10000],
    'LoanAmount': [100.0],
    'Loan_Amount_Term': [360.0],
    'Credit_History': [1.0],
    'Property_Area': ['Urban']
})

for col in encodings:
    new_data[col] = encodings[col].transform(new_data[col])
```
